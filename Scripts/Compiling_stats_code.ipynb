```python
import pandas as pd
```

```python

Stats1 = pd.read_csv("C:/Users/putna/Downloads/stats1(Sheet1).csv", encoding="ISO-8859-1")
Stats2 = pd.read_csv("C:/Users/putna/Downloads/stats2(Sheet1).csv", encoding="ISO-8859-1")
```

```python
print(Stats1.head())
print(Stats2.head())
```

```python
print(Stats1.columns)
print(Stats2.columns)
```

```python
# Select relevant columns from Stats1
stats1_filtered = Stats1[['Player', 'Team', 'Age', 'GP', 'Min', '+/-']]

# Select relevant columns from Stats2
stats2_filtered = Stats2[['Player', 'Age', 'Team', 'GS', 'MP', 'PER', 'WS', 'WS/48', 'VORP', 'Awards']]
```

```python
# Merge Stats1 and Stats2, keeping only players from Stats1
merged_stats = stats1_filtered.merge(stats2_filtered, on=['Player'], how='left')
```

```python
print(merged_stats.head())
```
